### Messing Values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

for feature in features_nan:
    print(f"{feature}: {np.round(df[feature].isnull().mean()*100,2)}%")
```

There are some feature which are more than 50% misssing.Its better to remove them from dataset beacause it could not contribute any thing.

```python

columns_to_drop = []

# Iterate over features with NaN values
for feature in features_nan:
    missing_percentage = np.round(df[feature].isnull().mean() * 100, 0)
    if missing_percentage >= 50:
        columns_to_drop.append(feature)  # Collect columns to drop

# Drop the identified columns
df = df.drop(columns=columns_to_drop)

# Print the cleaned dataset
df.head()
```

```python
# After removing the some feature
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

for feature in features_nan:
    print(f"{feature}: {np.round(df[feature].isnull().mean()*100,2)}%")
```

```python
## Replace missing value with a new label
def replace_cat_feature(df,features_nan):
    data=df.copy()
    data[features_nan]=data[features_nan].fillna('NA')
    return data

df=replace_cat_feature(df,features_nan)

df[features_nan].isnull().sum()
```

```python
df.head()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

for feature in numerical_nan:
    print(f"{feature}: {np.round(df[feature].isnull().mean()*100,2)}%")
```

```python
## Replacing the numerical Missing Values

for feature in numerical_nan:
    ## We will replace by using median since there are outliers
    median_value=df[feature].median()

    df[feature]=df[feature].fillna(median_value)

print(df[numerical_nan].isnull().sum())
```

### Handling Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
print(categorical_features)
```

```python
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
```

```python
df.head(10)
```

```python
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

## Model Building and Evaluation

In this part, we will build our prediction model

 we will follow these steps to build a model:

1. Choose an algorithm that implements the corresponding technique
2. Search for an effective parameter combination for the chosen algorithm
3. Create a model using the found parameters
4. Train (fit) the model on the training dataset
5. Test the model on the test dataset and get the results

### Splitting the Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=3)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

### Modeling

#### Linear Regression

##### Ridge Regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

parameter_space = {
    "alpha": [1, 10, 100, 290, 500],
    "fit_intercept": [True, False],
    "solver": ['svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga'],
}

clf = GridSearchCV(Ridge(random_state=3), parameter_space, n_jobs=4,
                   cv=3, scoring="neg_mean_absolute_error")

clf.fit(X_train, y_train)
print("Best parameters:")
print(clf.best_params_)
```

We defined the parameter space above using reasonable values for chosen parameters. Then we used GridSearchCV() with 3 folds (cv=3). Now we build our Ridge model with the best parameters found:

```python
ridge_model = Ridge(random_state=3, **clf.best_params_)
```

Then we train our model using our training set (X_train and y_train):

```python
ridge_model.fit(X_train, y_train);
```

Now we will evaluate our model On X_test by comparing actual value with predicted value using MAE(Mean Absolute Error)

```python
from sklearn.metrics import mean_absolute_error

y_pred=ridge_model.predict(X_test)
ridge_mae=mean_absolute_error(y_test,y_pred)

print(f"Ridge Regression MAE:{ridge_mae}")
```

##### Elastic Net

Here again,We will use `GridSearchCV` to find the best suited parameter of model.The parameter alpha is a constant that multiplies the penalty terms, l1_ratio determines the amount of L1 and L2 regularizations, fit_intercept is the same as Ridge's.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import ElasticNet

parameter_space={
    "alpha":[1,10,100,280,500],
    "l1_ratio": [0.5, 1],
    "fit_intercept":[True,False]
}

clf=GridSearchCV(ElasticNet(random_state=3),
                 parameter_space,
                 n_jobs=4,
                 cv=3,
                 scoring='neg_mean_absolute_error')
# a negative value because, by convention, GridSearchCV tries to maximize the score (higher is better).
# A higher (less negative) score indicates a better-performing model.
clf.fit(X_train,y_train)

print("Best Parameter:")
print(clf.best_params_)
```

```python
elasticNet_model=ElasticNet(random_state=3,**clf.best_params_)
```

```python
from sklearn.metrics import mean_absolute_error

elasticNet_model.fit(X_train,y_train)
y_pred=elasticNet_model.predict(X_test)

elasticNet_mae=mean_absolute_error(y_test,y_pred)

print("Elastic Net MAE :", elasticNet_mae)
```

#### Nearest Neighbors

We will use an implementation of the `k-nearest neighbors(KNN)` algorithm.

Firstly, we will use `GridSearchCV()` to search for the best model parameters in a parameter space provided by us.

```python
from sklearn.neighbors import KNeighborsRegressor

parameter_space={
    "n_neighbors":[9,10,11,50],
    "weights":['uniform','distance'],
    "algorithm":['ball_tree','kd_tree','brute'],
    "leaf_size":[1,2,5,20,50,200]
}

clf=GridSearchCV(KNeighborsRegressor(),
                 parameter_space,
                 n_jobs=4,
                 cv=3,
                 scoring='neg_mean_absolute_error')

clf.fit(X_train,y_train)
print('Best Parameter:')
print(clf.best_params_)
```

```python
knn_model=KNeighborsRegressor(**clf.best_params_)

knn_model.fit(X_train,y_train)
y_pred=knn_model.predict(X_test)
knn_mae=mean_absolute_error(y_test,y_pred)


print('KNN MAE:',knn_mae)
```

#### Suport Vector Regressor

For `Support Vector Regression(SVR)`, we will use one of three implementations provided by the Scikit-Learn package.

Firstly, we will use `RandomizedSearchCV()` to search for the best model parameters in a parameter space provided by us. The parameter `kernel` specifies the kernel type to be used in the algorithm, `degree` represents the degree of the polynomial kernel `poly`, `gamma` is the kernel coefficient for `rbf`, `poly` and `sigmoid` kernels, `coef0` is independent term in kernel function, and `C` is the penalty parameter of the error term.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVR

parameter_space={
    'kernel':['poly','rbf','linear','sigmoid'],
    'degree':[3,5],
    'gamma':[1e-3,1e-1,1/X_train.shape[1]],
    'C':[1,10,100]
}

clf = RandomizedSearchCV(SVR(),
                         parameter_space,
                         cv=3,
                         n_jobs=4,
                         scoring="neg_mean_absolute_error",
                         n_iter=100
                   )

clf.fit(X_train, y_train)
print("Best parameters:")
print(clf.best_params_)
```

```python
svr_model = SVR(**clf.best_params_)
svr_model.fit(X_train,y_train)
y_pred=svr_model.predict(X_test)

svr_mae=mean_absolute_error(y_test,y_pred)

print('SVR MAE:',svr_mae)
```
